```python
import torch
from torch.utils.data import DataLoader
from datasets import load_dataset,concatenate_datasets
from pyflink.datastream import StreamExecutionEnvironment
from pyflink.datastream.functions import MapFunction,SinkFunction
from pyflink.table import StreamTableEnvironment, EnvironmentSettings

# Assuming your model script is named classifier.py
from classifier import MultimodalSentimentClassifier
import time,csv,json
import PyCANDY as candy
# Step 1: Load IMDB dataset
def loadImdbDataset(trainSize=5000,testSize=500):
    dataset = load_dataset("imdb")
    
    combined_data = concatenate_datasets([dataset['train'], dataset['test']])
    # Shuffle the combined dataset
    combined_data = combined_data.shuffle(seed=999)

    # Ensure the requested sizes fit within the combined dataset size
    total_size = len(combined_data)
    if trainSize + testSize > total_size:
        raise ValueError("Requested train_size + test_size exceeds the total combined dataset size.")

    # Split into new train and test sets
    new_train_dataset = combined_data.select(range(trainSize))
    new_test_dataset = combined_data.select(range(trainSize, trainSize + testSize))

    return new_train_dataset, new_test_dataset
```
